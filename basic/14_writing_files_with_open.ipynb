<h1>Writing Files with Open</h1>

```python
# Create a file object with write mode
# 'w' parameter for write mode.

with open("resources/data/Example2.txt", "w") as file1:
    file1.write("This is line A\n")
    file1.write("This is line B\n")

# Print the result
with open("resources/data/Example2.txt", "r") as file1:
    file_content = file1.read()
    print(file_content)

# Using for loop to write
Lines = ["This is line 1\n", "This is line 2\n", "This is line 3\n"]
with open("resources/data/Example2.txt", "w") as file2:
    for line in Lines:
        file2.write(line)

# Print the result
with open("resources/data/Example2.txt", "r") as file2:
    file_content = file2.read()
    print(file_content)
```

```python
# Using 'a' parameter for append mode
# The mode will append content to the end of the file without overwriting existing content.
with open("resources/data/Example2.txt", "a") as file3:
    file3.write("This is line 4\n")
    file3.write("This is line 5\n")

# Print the result
with open("resources/data/Example2.txt", "r") as file3:
    file_content = file3.read()
    print(file_content)
```

```python
# Copy a file to another file
with open("resources/data/Example1.txt", "r") as readfile:
    with open("resources/data/Example3.txt", "w") as writefile:
        for line in readfile:
            writefile.write(line)
```
