```python
# Exception Handling
# Deal with errors

# try except block
try:
    getfile = open("myfile", "r")
    getfile.write("My file for exception handling")
except IOError: # Deal with error
    print("Unable to open or read the data in the file.")
else: # Notify system if there is no error
    print("The file was written sucessfully.")
finally: # Execute this block of code no matter what
    getfile.close()
    print("The file was closed successfully.")
```

```python
# Common Exceptions
# ZeroDivisionError
try:
    num1 = 10
    num2 = 0
    result = num1 / num2
except ZeroDivisionError:
    print("You cannot divide a number by zero.")

# ValueError
try:
    num = int("abc")
except ValueError:
    print("You cannot convert a string to an integer.")

# FileNotFoundError
try:
    getfile = open("nonexistentfile.txt", "r")
except FileNotFoundError:
    print("The file you are trying to open does not exist.")

# IndexError
try:
    mylist = [1, 2, 3]
    print(mylist[5])
except IndexError:
    print("You are trying to access an index that does not exist in the list.")

# KeyError
try:
    mydict = {"name": "John", "age": 30}
    print(mydict["gender"])
except KeyError:
    print("The key you are trying to access does not exist in the dictionary.")

# TypeError
try:
    num = 10 + "5"
except TypeError:
    print("You cannot add a number and a string together.")

# AttributeError
try:
    text = "Hello"
    length = len(text)
    missing = text.some_method()  # This method does not exist
except AttributeError:
    print("You are trying to access an attribute or method that does not exist.")

# ImportError
try:
    import non_existent_module
except ImportError:
    print("The module you are trying to import does not exist.")
```
