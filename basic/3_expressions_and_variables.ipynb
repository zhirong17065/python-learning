<h3>Arithmetic Operators</h3>

```python
# Expressions are combinations of operands and operators.
# Addition
a = 3 + 2
print(a, type(a))

# Subtraction
b = 5 - 3
print(b, type(b))

# Addition and Subtraction with integer and float values
c = 3 + 2.5
print(c, type(c)) # The result is a float because one of the operands is a float.

d = 5.0 - 3
print(d, type(d)) # The result is a float because one of the operands is a float.
```

```python
# Multiplication
e = 4 * 2
print(e, type(e))

# Division
f = 10 / 2
print(f, type(f))

# Floor Division
g = 10 // 3
print(g, type(g)) # The result is an integer because floor division returns the largest integer
```

<h3>Mathematical conventions</h3>

```python
# Mathematical conventions
# Python performs multiplication and division before addition and subtraction
h = 2 * 60 + 30
print("h =", h) # h = (2 * 60) + 30

i = 30 + 2 * 60
print("i =", i)  # i = 30 + (2 * 60)
```

```python
# Using parentheses to change the order of operations
j = (30 + 2) * 60
print("j =", j) # j = (30 + 2) * 60
```

<h2>Variables</h2>

```python
# '=' is the assignment operator.
my_variable = 10 # Assigning a value '10' to a variable 'my_variable'

# Assigning a new value to the same variable
my_variable = 20
```

```python
# Store the result of an expression in a variable
x = 43 + 60 + 16 + 41
print("x =", x) # x = 160

# Perform operations on x
y = x / 60
print("y =", y) 

# Perform operations on x and store the result in x
x = x / 60
print("x =", x) 
```
