<h2> REST APIs</h2>
<p>REST (Representational State Transfer) APIs allows communication between client and server over HTTP.</p>
<p>HTTP is a protocol for transferring data over the internet. Client sends requests to the server, and the server responds with the requested data via HTTP, usually containing a JSON file. After processing, the server returns the response to the client, usually also using a JSON file.</p>

<h2>PyCoinGecko for CoinGecko API</h2>
<p>PyCoinGecko is a Python wrapper for the CoinGecko API, which provides access to cryptocurrency data.</p>

```python
# install API package
!pip3 install pycoingecko
```

```python
from pycoingecko import CoinGeckoAPI
# API object
cg = CoinGeckoAPI()

# Using API to retrieve data
bitcoin_data = cg.get_coin_market_chart_by_id(id = 'bitcoin', vs_currency = 'usd', days = 30)
bitcoin_data['prices']
```

```python
# Convert the data into a pandas DataFrame
import pandas as pd
df = pd.DataFrame(bitcoin_data['prices'], columns=['TimeStamp', 'Price'])
# Convert the timestamp from milliseconds to a datetime object
df['Date'] = pd.to_datetime(df['TimeStamp'], unit='ms')
df
```

```python
# Create a candlestick plot
candlestick_data = df.groupby(df.Date.dt.date).agg({'Price':['min', 'max', 'first', 'last']})
candlestick_data
```

```python
!pip3 install plotly
!pip3 install nbformat
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Candlestick(x=candlestick_data.index,
                open=candlestick_data['Price']['first'],
                high=candlestick_data['Price']['max'],
                low=candlestick_data['Price']['min'],
                close=candlestick_data['Price']['last'])]
                )

fig.update_layout(
    xaxis_rangeslider_visible=False, 
    xaxis_title='Date', 
    yaxis_title='Price (USD $)',
    title='Bitcoin Price Candlestick Chart Over Past 30 Days'
)
```
