<h1>Part A</h1>
<p>Creating a class</p>

<h2>Step 1: Define a string</h2>

```python
givenstring = "Lorem ipsum dolor! diam amet, consetetur Lorem magna. sed diam nonumy eirmod tempor. diam et labore? et diam magna. et diam amet."
```

<h2>Step 2: Define the class and its attributes</h2>
<h2>Step 3: Implement a code to format the text in lowercase</h2>
<h2>Step 4: Implement a code to count the frequency of all unique words</h2>
<h2>Step 5: Implement a code to count the frequency of a specific word</h2>

```python
class TextAnalyzer:
    '''
    A class that analyzes a given text
    '''
    fmtText = ""
    punctuation_marks = [".", ",", "!", "?"]
    def __init__(self, text):
        '''
        Initialize the TextAnalyzer with the given text
        Turn the text to lowercase and remove punctuation marks
        input: text (str): The text to be analyzed
        '''
        # Turn the text to lowercase
        self.fmtText = text.lower()

        # Remove punctuation marks from the text
        for mark in self.punctuation_marks:
            self.fmtText = self.fmtText.replace(mark, "")
    
    def freqAll(self):
        '''
        The function to count the frequency of all unique words in the text
        1. Split the fmtText attribute into individual words
        2. Create an empty dictionary to store the word frequency
        3. Iterate over the list of words and update the frequency dictionary accordingly
        4. Count the occurence
        5. Return the frequency dictionary
        '''
        words = self.fmtText.split()
        freq_dict = {}
        for word in words:
            if not(word in freq_dict):
                freq_dict[word] = words.count(word)
        return freq_dict
    
    def freqOf(self, word):
        '''
        The function to count the frequency of a specific word
        input: word (str): The word to be counted
        output: freq (int): The frequency of the word in the fmtText
        '''
        freq_dict = self.freqAll()
        if word in freq_dict:
            return freq_dict[word]
        else:
            return 0
```

<h1>Part B</h1>
<p>Call the functions created in Part A, allowing the functions to execute and generate output.</p>

<h2>Step 1: Create an instance of TextAnalyzer class</h2>

```python
text_analyzer = TextAnalyzer(givenstring)
```

<h2>Step 2: Call the function that converts the data into lowercase</h2>

```python
print("Formatted Text:", text_analyzer.fmtText)
```

<h2>Step3: Call the function that count the frequency of all unique words from the data</h2>

```python
text_analyzer.freqAll()
```

<h2>Step 4: Call the function that counts the frequency of a specific word</h2>

```python
text_analyzer.freqOf("lorem")
```
