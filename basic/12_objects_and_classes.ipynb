```python
# Classes
# A class is a blueprint for creating objects

import matplotlib.pyplot as plt
class Shape:
    # Attributes
    color = None
    area = None

class Circle(Shape): # Circle is a subclass of Shape
    # Attributes
    radius = None

    # Constructor
    def __init__(self, color, radius):
        self.color = color
        self.radius = radius
    
    # Destructor
    def __del__(self):
        print("Delete Circle object")

    # Method
    def add_radius(self, r):
        self.radius += r
        print("New radius is", self.radius)
        pass
    
    def draw_circle(self):
        # Draw the circle
        plt.gca().add_patch(plt.Circle((0, 0), self.radius, color=self.color))
        plt.axis('scaled')
        plt.show()
        pass



class Rectangle(Shape):
    # Attributes
    height = None
    width = None

    # Constructor
    def __init__(self, color, height, width):
        self.color = color
        self.height = height
        self.width = width
    
    # Destructor
    def __del__(self):
        print("Delete Rectangle object")

# Crearte instances of Circle and Rectangle
c1 = Circle("red", 5)
c2 = Circle("blue", 10)
r1 = Rectangle("green", 4, 6)
r2 = Rectangle("yellow", 8, 12)
c1.draw_circle()

# Access attributes of Circle and Rectangle
print(c1.radius)
print(r1.width)

```
