```python
import pandas as pd
df = pd.read_csv("resources/csv/songs.csv", sep=';')
df.head()
```

```python
# Unique value dataframe
unique_df = df['Released'].unique()
unique_df
```

```python
# Appying inequality operators for filtering
print(df['Released'] > 1979) # return a boolean series
filtered_df = df[df['Released'] > 1979]
filtered_df
```

```python
# save to csv
filtered_df.to_csv("resources/csv/filtered_songs.csv")
```
