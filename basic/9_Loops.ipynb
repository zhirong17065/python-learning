```python
# Loops
# range() function generates a sequence of numbers
# range(N) -> [0, 1, 2, ..., N-1]
for i in range(3):
    print(i)

# range(start, stop) -> [start, ..., stop-1]
for i in range(10, 15):
    print(i)
```

```python
# For loops
squares = ["red", "yellow", "green", "purple", "blue"]

# Using index
for i in range(0,5):
    squares[i] = "white"
print(squares)

# Using object reference
squares = ["red", "yellow", "green", "purple", "blue"]
for square in squares:
    square = "white"
```

```python
# Using enumerate() to get both index and object reference
squares = ["red", "yellow", "green"]
for i, square in enumerate(squares):
    print(i, square)
```

```python
# while loops
squares = ["orange", "orange", "purple", "orange", "blue"]
New_squares = []
i = 0
while squares[i] == "orange":
    New_squares.append(squares[i])
    i += 1

print(New_squares)
```

```python
# Break and continue statements
for num in range(10):
    if num == 5:
        print("Breaking the loop at:", num)
        break
    print(num)

for num in range(10):
    if num == 5:
        print("Skipping the loop at:", num)
        continue
    print(num)
```
