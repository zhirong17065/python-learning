```python
# Strings
# Use single or double quotes to represent a string
a = "Hello"
b = 'Hello'

# Strings can contain numbers, special characters, and spaces
c = "Hello 123!@#"
```

```python
# Consider strings as a sequence of characters. Access individual characters using indexing
name = "Michael Jackson"
print("The first character is:", name[0])  # Output: M
print("The seventh character is:", name[6])  # Output: l

# Negative indexing allows you to access characters from the end of the string
print("The last character is:", name[-1])  # Output: n
print("The fifteenth character from the end is:", name[-15])  # Output: M

# String slicing
# str[start:end] returns a substring from index start to end-1
print("The first seven characters are:", name[0:7])  # Output: Michael
print("The last seven characters are:", name[8:15])  # Output: Jackson

# String striding
print("Every second character from the string is:", name[::2])  # Selects every second character from the string
print("Every second character from index 0 to 4 is:", name[0:5:2]) # Selects every second character from index 0 to 4

# Length of a string
print("The length of the string is:", len(name))  # Output: 15
```

```python
# String concatenation
statement = name + "is the best"
print(statement)  # Output: Michael Jackson is the best

# Multiplying strings
# Results in the string being repeated a certain number of times
name = "Michael Jackson "
print(name * 3)
```

```python
# String is immutable
# Characters in a string cannot be changed after the string is created
name[0] = "m"  # This will raise an error
```

```python
# Using new assignment to change the string
name = name + "is the best"
print(name)  
```

```python
# Escape sequences
# Newline
print("Michael Jackson\nis the best")

# Tab
print("Michael Jackson\tis the best")

# Using backslash to escape special characters
print("Michael Jackson\\is the best")

# Using r before the string to disable escape sequences
print(r"Michael Jackson\nis the best")
```

<h2>String Methods</h2>

```python
# String methods
A = "Michael Jackson is the best"

B = A.upper()  # Converts all characters to uppercase
print(B)

C = A.lower()  # Converts all characters to lowercase
print(C)

D = A.replace("Michael", "Janet")  # Replaces "Michael" with "Janet"
print(D)

E = A.find("el")  # Finds the index of the first occurrence of "el"
print(E)
F = A.find("error") # Returns -1 if the substring is not found
print(F)
```
