<h1>REST APIs & HTTP Requests</h1>

<h2>HTTP Protocol</h2>
<p>HTTP is the general protocol for transferring data over the internet.</p>
<p>Requests are sent when calling an REST API function. The request is communicated via HTTP message, usually containing a JSON file.</p>

<h2>URL</h2>
<p>Uniform Resource Locator (URL) consists 3 main parts:</p>
<ol>
<li>Scheme: The protocol</li>
<li>Internet address/Base URL</li>
<li>Route: location on the web server</li>
</ol>

```python
import requests
url = 'https://www.ibm.com'
r = requests.get(url)
print(r.status_code)
print(r.headers) # Return a dictionary of the response headers
print(r.request.body)
```

```python
header = r.headers
print(header['date'])
print(header['content-type'])
print(r.encoding)
print(r.text[0:100])
```

<h2>Get request with URL parameters</h2>
<ul>
<li>Base URL: httpbin.org</li>
<li>Route: /get</li>
</ul>
<p>Exapmle: httpbin.org/get</p>

<h3>Query string</h3>
<ul>
<li>Start of query: ?</li>
<li>Parameter Name: name</li>
<li>=</li>
<li>Value: Joseph</li>
<li>&</li>
<li>Parameter Name: ID</li>
<li>=</li>
<li>Value: 123</li>
</ul>
<p>Example: ?name=Joseph&ID=123</p>


```python
url_get = 'http://httpbin.org/get'
payload = {"name":"Joseph", "ID":"123"}
r = requests.get(url_get, params=payload)
print(r.url)
print(r.request.body)
```

```python
print(r.text)
print(r.headers['Content-Type'])
```

```python
# format the content type using JSON
r.json()
r.json()['args']
```

<h2>POST Request</h2>

```python
url_post = 'http://httpbin.org/post'
payload = {"name":"Joseph", "ID":"123"}
r_post = requests.post(url_post, data=payload)
```
