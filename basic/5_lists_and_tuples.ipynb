```python
# Tuple is an ordered sequence of elements that is immutable
tuple1 = (1, 2, 3, 4, 5)

# Different types of elements can be stored in a tuple
tuple2 = (1, "Hello", 3.14, True)
print(type(tuple2))

# Access using index
for i in range(len(tuple2)):
    print("The ", i, " element is: ", tuple2[i])

# Access using negative index
for i in range(-1, -len(tuple2)-1, -1):
    print("The ", i, " element is: ", tuple2[i])
```

```python
# Combine tuples using + operator
tuple3 = tuple2 + ("hard rock", 10)
print(tuple3)
```

```python
# Slicing a tuple
# new tuple = tuple[start:end] slices the tuple from index start to end-1

tuple4 = tuple3[1:4] # Slicing from index 1 to 3
print(tuple4)
```

```python
# Tuple is immutable
# Assign a new object to the tuple variable to change its value
ratings = (10, 6, 8, 9, 5)
sorted_ratings = sorted(ratings) # create and assign a new sorted list to variable
print("Original ratings: ", ratings)
print("Sorted ratings: ", sorted_ratings)
```

```python
# Nesting
# A tuple can contain other tuples as elements
NT = (1, 2, ("rock", "pop"), (3, 4), ("disco", (1, 2)))

# Access using double indexing
print(NT[2][0]) # Accessing "rock"

# Accessing character in a string inside a tuple
print(NT[2][0][3]) # Accessing "k" in "rock
```

```python
# Lists
# List is a mutable ordered sequence of elements
L = ["Michael Jackson", 10.1, 1982]

# List can contain different types of elements, like strings, floats, integers, even tuples, and other lists
L1 = ["Michael Jackson", 10.1, 1982, [1, 2], ('A', 1)]

# Access using index
print("The first element is: ", L[0])

# Access using negative index
print("The last element is: ", L[-1])

# Slicing a list
# new list = list[start:end] slices the list from index start to end-1
L2 = L[0:3] # Slicing from index 0 to 2
print(L2)

# Concatenate lists using + operator
L3 = L + ["pop", 10]
print(L3)
```

```python
# List is mutable
# Extend a list with another list using extend() method
L = ["Michael Jackson", 10.1, 1982]
L.extend(["pop", 10])
print(L)

# Append an element to the end of the list using append() method
L.append("A")
print(L)

# Changing an element in the list using index
A = ["disco", 10, 1.2]
A[0] = "hard rock"
print(A)

# Using del() to delete an element from the list
del(A[1])
print(A)

# Convert a string to a list using split() method
s1 = "Michael Jackson"
s1_list = s1.split()
print(s1_list)

# Split by a specific character (delimiter)
s2_list = "A,B,C,D".split(",")
print(s2_list)
```

```python
# Aliasing
# Assigning a list to another variable creates an alias, not a copy
A = ["hard rock", 10, 1.2]
B = A

# Both A and B point to the same list object
B[0] = "banana"
print("A: ", A)
print("B: ", B)
```

```python
# Clone a list to avoid aliasing
A = ["hard rock", 10, 1.2]
B = A[:]

# Now A and B point to different list objects
A[0] = "banana"
print("A: ", A)
print("B: ", B)

help(A)
```
