<h1>All mode of open function</h1>

```python
# r
# read mode
# Opens an existing file for reading. 
# Raises an error if the file doesn't exist.
with open("resources/data/Example1.txt", "r") as file1:
    content = file1.read()
    print(content)
```

```python
# w 
# write mode
# Creates a new file for writing. 
# Overwrites the file if it already exists.
with open("resources/data/Example2.txt", "w") as file2:
    file2.write("This is line 1\n")
    file2.write("This is line 2\n")
    file2.write("This is line 3\n")
```

```python
# a
# append mode
# Opens a file for appending data. 
# Creates the file if it doesn't exist.
with open("resources/data/Example2.txt", "a") as file2:
    file2.write("This is an appended line 1\n")
    file2.write("This is an appended line 2\n")
```

```python
# x
# exclusive creation mode
# Creates a new file for writing but raises an error if the file already exists.
with open("resources/data/Example3.txt", "x") as file3:
    file3.write("This is line 1\n")
    file3.write("This is line 2\n")
```

```python
# rb
# read binary mode
# Opens an existing binary file for reading.
with open("resources/data/Example1.txt", "rb") as file4:
    binary_content = file4.read()
    print(binary_content)

# wb
# write binary mode
# Creates a new binary file for writing.
with open("resources/data/Example4.bin", "wb") as file5:
    file5.write(b"This is binary data\n")
    file5.write(b"This is more binary data\n")

# ab
# append binary mode
# Opens a binary file for appending data.
with open("resources/data/Example4.bin", "ab") as file5:
    file5.write(b"This is appended binary data\n")

# xb
# exclusive creation binary mode
# Creates a new binary file for writing but raises an error if the file already exists.
with open("resources/data/Example5.bin", "xb") as file6:
    file6.write(b"This is binary data\n")
    file6.write(b"This is more binary data\n")
```

```python
# rt
# read text mode
# Opens an existing text file for reading.
with open("resources/data/Example1.txt", "rt") as file7:
    text_content = file7.read()
    print(text_content)

# wt
# write text mode
# Creates a new text file for writing.
with open("resources/data/Example6.txt", "wt") as file8:
    file8.write("This is line 1\n")
    file8.write("This is line 2\n")
    file8.write("This is line 3\n")

# at
# append text mode
# Opens a text file for appending data.
with open("resources/data/Example6.txt", "at") as file8:
    file8.write("This is an appended line 1\n")
    file8.write("This is an appended line 2\n")

# xt
# exclusive creation text mode
# Creates a new text file for writing but raises an error if the file already exists.
with open("resources/data/Example7.txt", "xt") as file9:
    file9.write("This is line 1\n")
    file9.write("This is line 2\n")
```

```python
# r+
# read and write mode
# Opens an existing file for both reading and writing.
with open("resources/data/Example1.txt", "r+") as file10:
    content = file10.read()
    print(content)
    file10.write("This is an additional line\n")

# w+
# write and read mode
# Creates a new file for writing and reading.
# Overwrites the file if it already exists.
with open("resources/data/Example8.txt", "w+") as file11:
    file11.write("This is line 1\n")
    file11.write("This is line 2\n")
    file11.write("This is line 3\n")
    file11.seek(0)  # Move the cursor to the beginning of the file
    content = file11.read()
    print(content)

# a+
# append and read mode
# Opens a file for both appending and reading.
# Creates the file if it doesn't exist.
with open("resources/data/Example8.txt", "a+") as file11:
    file11.write("This is an appended line 1\n")
    file11.write("This is an appended line 2\n")
    file11.seek(0)  # Move the cursor to the beginning of the file
    content = file11.read()
    print(content)

# x+
# exclusive creation and read mode
# Creates a new file for writing and reading but raises an error if the file already exists.
with open("resources/data/Example9.txt", "x+") as file12:
    file12.write("This is line 1\n")
    file12.write("This is line 2\n")
    file12.seek(0)  # Move the cursor to the beginning of the file
    content = file12.read()
    print(content)
```
