```python
# Functions
# Define a function
def f1(input):
    """add 1 to input"""
    output = input + 1
    return output

# Call the function
result = f1(1)
print(result)
```

```python
# Python build-in functions
# len()
# input: type sequence
# output: type int - length of the sequence 
album_ratings = [10.0, 8.5, 9.0, 8.0, 7.5]
L = len(album_ratings)
print(L)

# sum()
# input: type sequence
# output: type int or float - sum of the sequence
S = sum(album_ratings)
print(S)
```

```python
# Functions for sorting
# sorted()
# input: type sequence
# output: type list - sorted list of the sequence
album_ratings = [10.0, 8.5, 9.0, 8.0, 7.5]
sorted_album_ratings = sorted(album_ratings)
print(album_ratings)
print(sorted_album_ratings)

# sort()
# input: type list
# output: type None - sort the list in place
album_ratings = [10.0, 8.5, 9.0, 8.0, 7.5]
album_ratings.sort()
print(album_ratings)
```

```python
# Using the keyword pass to define an empty function
def empty_function():
    pass

print(empty_function()) # None
```

```python
# Using * to define a function with variable number of arguments
def artistNames(*names):
    for name in names:
        print(name)

artistNames("Michael Jackson", "Madonna", "Prince")
```

```python
# Scope
def AddDC(x):
    x = x + "DC" # function scope
    print(x)
    return x

x = "AC" # global scope
z = AddDC(x)

# Function scope has higher priority than global scope, so the output of the function is "ACDC" and the value of z is also "ACDC". However, the value of x in the global scope remains unchanged as "AC".
def Thriller():
    Date = "1982" 
    return Date

Date = "2017"
print(Thriller()) # Read the functino scope variable
print(Date) # Read the global scope variable
```

```python
# global keyword
def PinkFloyd():
    global ClaimedSales
    ClaimedSales = '45 million'
    return ClaimedSales

PinkFloyd()
print(ClaimedSales)
```
