<h1>Reading Files with Open Function</h1>

```python
# Read a file object

# open() function is used to open a file in Python.
# parameter 1: file path, consist of the directory and file name
# parameter 2: mode
# return: file object
file1 = open("resources/data/Example1.txt", "r") # 'r' for read mode

# Access file object
print(file1.name)
print(file1.mode)

# Read the file content
file_content = file1.read()
print(file_content)

# Alway close the file after use
file1.close()

# Check if the file is closed
print(file1.closed)
```

```python
# Using with statement is a better pratice
# Its automatically closes the file
with open("resources/data/Example1.txt", "r") as file2:
    file_content = file2.read()
    print(file_content)
print(file2.closed)
print(file_content)
```

```python
# readlines() method
# return a list of lines in the file
with open("resources/data/Example1.txt", "r") as file3:
    file_content = file3.readlines()
    print(file_content)

# readline() method
# return a single line from the file
# second call will return the next line
# taking an argument n will read n bytes from the line
with open("resources/data/Example1.txt", "r") as file4:
    file_content = file4.readline()
    print(file_content)
    file_content = file4.readline()
    print(file_content)
    file_content = file4.readline(5) # read next 5 bytes from the line
    print(file_content)
    file_content = file4.readline(5) # read next 5 bytes from the line
    print(file_content)
```
