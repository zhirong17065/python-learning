```python
# Sets
# Sets are type of collection
# The elements is unordered, unique, and immutable

set1 = {"pop", "rock", "soul", "hard rock", "rock", "R&B", "rock", "disco"}
print(set1) # Duplicates are removed
```

```python
# Using set() to convert a list to a set
album_list = ["Michael Jackson", "Thriller", "Thriller", 1982]
album_set = set(album_list)
print(album_set)
```

```python
# Set Operations
# Consider a set in a venn diagram
A = {"Thriller", "Back in Black", "AC/DC"}

# Add an element to a set
A.add("NSYNC")
print(A)
A.add("NSYNC") # Add dulicate element
print(A) # No change to the set

# Remove an element from a set
A.remove("NSYNC")
print(A)

# Verifying membership in a set
print("AC/DC" in A) # Return true if the element is in the set
print("NSYNC" in A) # Return false if the element is not in the set
```

```python
# Set Mathematical Operations
album_set_1 = {"AC/DC", "Back in Black", "Thriller"}
album_set_2 = {"AC/DC", "Back in Black", "The Dark Side of the Moon"}

# Using & (AND) to find the intersection of two sets
album_set_3 = album_set_1 & album_set_2
print(album_set_3) # Only save the elements that are in both sets

# Using | (OR) to find the union of two sets
album_set_4 = album_set_1 | album_set_2
print(album_set_4) # Save all the elements in both sets

# Using - (MINUS) to find the difference of two sets
album_set_5 = album_set_1 - album_set_2
print(album_set_5) 

# Using issubset() to check is a set is a subset of another set
print(album_set_3.issubset(album_set_1)) # Return True if album_set_3 is a subset of album_set_1

# Using issuperset() to check is a set is a superset of another set
print(album_set_1.issuperset(album_set_3)) # Return True if album_set_1 is a superset of album_set_3
```
