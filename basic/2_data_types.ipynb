```python
# Data Types in Python
# int: 1
# float: 1.0
# str: "Hello"

# Using type() function to check the data type of value
print(type(1))
print(type(1.0))
print(type("Hello"))
```

```python
# Integer
# Discrete whole numbers, can be positive or negative, and have no decimal point.
# The range of integers is large, but it is limited by the amount of memory available on the system.

x = 10
print(x, type(x))

y = -5
print(y, type(y))

import sys
print(sys.int_info) # Checking the range and precision of integer values in Python
```

```python
# Float
# Continuous numbers that can have a decimal point. They can also be positive or negative.
# Consider the value between 1 and 2, which can be represented as 1.0, 1.1, 1.2, ..., 1.9, 2.0.

x = 3.14
print(x, type(x))

y = -0.001
print(y, type(y))

import sys
print(sys.float_info) # Checking the precision and range of float values in Python
```

```python
# Typecasting
# Converting one data type to another data type.

a = 10  # Integer
b = float(a)  # Convert integer to float
print(b, type(b))

c = 3.14  # Float
d = int(c)  # Convert float to integer
print(d, type(d)) # The float value is truncated, not rounded.

e = "100"  # String
f = int(e)  # Convert string to integer
print(f, type(f))

g = 1 # Integer
h = str(g) # Convert integer to string
print(h, type(h))

i = 2.718 # Float
j = str(i) # Convert float to string
print(j, type(j))

k = "A" # String
l = int(k) # This will raise a ValueError because "A" is not a integer string.
```

```python
# Data type conversion after performing arithmetic operations
a = 6/2
print(type(a)) # The result of division is always a float.

b = 5//2
print(type(b)) # The result of floor division is an integer.

c = 5%2
print(type(c)) # The result of modulus operation is an integer.

d = "Hello, World!" == "Hello, World!"
print(type(d)) # The result of comparison operation is a boolean.
```

```python
# Boolean
# A boolean value can be either True or False. 

x = True
print(x, type(x))

y = False
print(y, type(y))
```

```python
# Typecasting of boolean values
w = True
x = int(w)  # Convert boolean to integer
print(x, type(x))  # True is converted to 1

y = False
z = int(y)  # Convert boolean to integer
print(z, type(z))  # False is converted to 0    

a = 1  # Integer
b = bool(a)  # Convert integer to boolean
print(b, type(b))  # Non-zero integers are converted to True

c = 0  # Integer
d = bool(c)  # Convert integer to boolean
print(d, type(d))  # Zero is converted to False
```
