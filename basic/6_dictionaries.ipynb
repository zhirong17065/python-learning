```python
# Dictionaries
# Each element in dictionary is a key-value pair
# The key must be unique and immutable
# The value can be immutable, mutable, or duplicate

dict_1= {"key1":1, "key2":"2", "key3":[3,3,3], "key4":(4,4,4), "key5":5}
```

```python
# Using key to access value
books = {"Thriller":1982,
        "Back in Black":1980, 
        "The Dark Side of the Moon":1973, 
        "The Bodyguard":1992, 
        "Bat Out of Hell":1977, 
        "Their Greatest Hits (1971-1975)":1976, 
        "Saturday Night Fever":1977, 
        "Rumours":1977}

# Return a value with a key
print(books["Thriller"])
print(books["Back in Black"])
```

```python
# Adding a new element to the dictionary
books["Graduation"] = 2007
print(books)

# Delete an element from the dictionary
del books['Thriller']
print(books)

# Verify if a key exists in the dictionary
if "The Bodyguard" in books: # Return a true if the key exists
    print("The Bodyguard is in the dictionary")
```

```python
# Get all the keys in the dictionary
titles = books.keys()
print(titles)

# Get all the values in the dictionary
years = books.values()
print(years)

# Use list() to convert the keys and values to a list
titles_list = list(books.keys())
print(titles_list)  
years_list = list(books.values())
print(years_list)
```
